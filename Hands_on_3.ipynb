 Timing the Function and Polynomial Fitting

```python
import time
import numpy as np
import matplotlib.pyplot as plt

# Original function to calculate runtime
def f(n):
    x = 1
    for i in range(1, n+1):
        for j in range(1, n+1):
            x += 1
    return x

# Time the function for various values of n
n_values = list(range(1, 501))  # Testing n from 1 to 500 for timing
times = []

for n in n_values:
    start_time = time.time()
    f(n)
    end_time = time.time()
    times.append(end_time - start_time)

# Plotting time vs n
plt.figure(figsize=(10, 6))
plt.plot(n_values, times, label="Execution Time")
plt.xlabel("n")
plt.ylabel("Time (seconds)")
plt.title("Time vs n")
plt.grid(True)
plt.show()
```

```python
import time
import numpy as np
import matplotlib.pyplot as plt

# Original function to calculate runtime
def f(n):
    x = 1
    for i in range(1, n+1):
        for j in range(1, n+1):
            x += 1
    return x

# Time the function for various values of n
n_values = list(range(1, 501))  # Testing n from 1 to 500 for timing
times = []

for n in n_values:
    start_time = time.time()
    f(n)
    end_time = time.time()
    times.append(end_time - start_time)

# Polynomial fitting using NumPy (replace with your actual fitting method)
coefficients = np.polyfit(n_values, times, 2) # Example: Fit a 2nd-degree polynomial

# Generate fitted values for all n_values
fit_times = [np.polyval(coefficients, n) for n in n_values] # Example using NumPy

# Zooming in to find the approximate n_0 where deviation occurs
plt.figure(figsize=(10, 6))
plt.plot(n_values, times, label="Original Times", color='blue')
plt.plot(n_values, fit_times, label=f"Polynomial Fit", color='red')
plt.xlim(0, 100)  # Zoom into n values between 0 and 100
plt.xlabel("n")
plt.ylabel("Time (seconds)")
plt.title("Zoomed Time vs n (0 to 100)")
plt.legend()
plt.grid(True)
plt.show()
```
